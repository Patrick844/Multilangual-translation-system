```python
from lingowiz.utils import model_dict
from lingowiz import metrics
from transformers import MarianTokenizer, MarianMTModel
from google.colab import userdata

token = userdata.get("HF_TOKEN")

tokenizer = MarianTokenizer.from_pretrained(model_dict["eng"][0], token=token)
model = MarianMTModel.from_pretrained(model_dict["eng"][0],token=token)
```

```python
import pandas as pd
import mlflow
mlflow.set_tracking_uri("https://40f6-34-16-251-104.ngrok-free.app/")
df = pd.read_csv("health_test.csv")
df = df.iloc[:1000]
x = metrics.evaluation(df,model,tokenizer,"English","Arabic")
df.iloc[0,0]
```
